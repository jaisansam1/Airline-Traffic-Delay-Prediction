```python
#checking outliers
df.plot(kind='box', subplots=True, layout=(5,15),
sharex=False,sharey=False, figsize=(20,50), 
color='red')
```

```python
for each_feature in df.columns[5:-5]:
    Q1 = df[each_feature].quantile(0.25) # 25th percentile of the data of the given feature
    Q2 = df[each_feature].quantile(0.50)
    Q3 = df[each_feature].quantile(0.75) # 75th percentile of the data of the given feature
    IQR = Q3-Q1 #Interquartile Range
    outlier_step = IQR * 1.5 
    outmask = (df[each_feature] > Q3 + outlier_step) | (df[each_feature] < Q1 - outlier_step)
    outdata = df[each_feature][outmask]
    percentile25 = df[each_feature].quantile(0.25)
    percentile75 = df[each_feature].quantile(0.75)
    upper_limit = percentile75 + 1.5 * IQR
    lower_limit = percentile25 - 1.5 * IQR
    df[each_feature] = df[each_feature].clip(upper_limit,lower_limit)
        
```
